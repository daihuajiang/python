```python
#圖片轉陣列函式
def image_to_array(filepath):
    result = np.array([])  # 建立一個空的一維陣列
    #print("開始將圖片轉為陣列")
    for i in range(img_num):
        image = Image.open(filepath[i])
        """
        r, g, b = image.split()  # rgb通道分離
        # 注意：下面一定要reshpae(image_width*image_hight)使其變為一維陣列，否則拼接的資料會出現錯誤，導致無法恢復圖片
        r_arr = np.array(r).reshape(image_width*image_hight)
        g_arr = np.array(g).reshape(image_width*image_hight)
        b_arr = np.array(b).reshape(image_width*image_hight)
        # 行拼接，共52列，一列image_width*image_hight行，為一張圖片的rgb值
        image_arr = np.concatenate((r_arr, g_arr, b_arr))
        result = np.concatenate((result, image_arr))
        """
        
        image_arr = image.convert('L')
        im = np.array(image_arr).reshape(image_width*image_hight)
        result = np.concatenate((result, im))
        
    
    result = result.reshape((img_num, rehape_size))  # 將一維陣列轉化為count列image_width*image_hight行的二維陣列
    return result
```
